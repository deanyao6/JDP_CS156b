```python
import os
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt

train_csv = "/resnick/groups/CS156b/from_central/data/student_labels/train2023.csv"
image_base = "/resnick/groups/CS156b/from_central/data"

df = pd.read_csv(train_csv)

first5 = df.head(20)

fig, axes = plt.subplots(1, 20, figsize=(100, 25))

for ax, (_, row) in zip(axes, first5.iterrows()):
    rel_path = row["Path"]                      # e.g. train/pid.../view1_frontal.jpg
    full_path = os.path.join(image_base, rel_path)

    img = Image.open(full_path)
    ax.imshow(img, cmap="gray")
    ax.set_title(
        f'{os.path.basename(rel_path)}\n{row["Sex"]}, age {row["Age"]}',
        fontsize=9
    )
    ax.axis("off")

plt.tight_layout()
plt.show()
```
